# Neural Networks Basics

## Core Components of a Neural Network
Three main things go into neural networks:
- **Input**
- **Output**
- **Target**


## Loss Function: Mean Squared Error (MSE)

Mean Squared Error (MSE) is also called the **loss**.

MSE formula (single value):
MSE = (output − target)²

Where:
- w = weight  
- b = bias  
- y = target  


## Simple Neural Network (Single Input Example)

A very basic neural network can be written as:

ŷ = wx + b

### Given
- x = 2 (input)  
- w = 3 (weight)  
- b = 1 (bias)  
- y = 10 (target)

### Output
ŷ = (3 × 2) + 1 = 7

### Residual (Error)
r = ŷ − y  
r = 7 − 10 = −3

### Mean Squared Error
MSE = (ŷ − y)²  
MSE = (−3)² = 9

## Sending More Than One Number Through the Network

When more than one value is sent through the network:
- **Vectors** are used
- **Outputs** are vectors
- **Targets** must also be vectors

Notation:
- X = input vector  
- Ŷ = target vector  
- Y = output vector 

## Residual Vector

The **residual** (r) is the difference between the target and the output:

r = Ŷ − Y

## Vector Example

### Given

Output vector:
Y =
[ 2  
  4  
  6 ]

Target vector:
Ŷ =
[ 3  
  5  
  4 ]

## Residual Calculation

r =
[ 3 − 2  
  5 − 4  
  4 − 6 ]

r =
[ 1  
  1  
 −2 ]

## Squared Residuals

r² =
[ 1²  
  1²  
 (−2)² ]

r² =
[ 1  
  1  
  4 ]

## Sum of Squared Errors (SSE)

Sum of Squared Errors is computed using the transpose:

SSE = rᵀ r

rᵀ = [ 1  1  −2 ]

SSE = (1 × 1) + (1 × 1) + (−2 × −2)  
SSE = 6

## Mean Squared Error (MSE) for Vectors

MSE is the **average** of the squared errors:

MSE = (1 / n) × SSE  
n = number of elements in the vector

For this example:
MSE = 6 / 3 = 2

## Formula Summary

- Output (single input):  
  ŷ = wx + b

- Residual:  
  r = Ŷ − Y

- Sum of Squared Errors:  
  SSE = rᵀ r

- Mean Squared Error (Loss):  
  MSE = (1 / n) rᵀ r

# Function With More Than One Dimension

f(x, y) = ax + by + c

- x, y are inputs  
- a, b are weights (one per input)  
- c is the bias  
- f(x, y) is the output

---

## Key Ideas

- More inputs means **more weights**
- Each input has its **own weight**
- The bias shifts the output up or down
- This is still a **linear function**

---

## Vector Form

Inputs:
X = [x, y]

Weights:
W = [a, b]

Output:
ŷ = WᵀX + c

---

## Neural Network Interpretation

- 2 input neurons  
- 1 output neuron  
- Each input connects to the output with a weight  
- A bias is added before producing the output  

This is the basic structure of a **single-layer neural network**.
